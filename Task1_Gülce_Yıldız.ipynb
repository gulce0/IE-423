# Fitting (Train) a Regression Model to Predict the Value of Football Players

```python
# Assign dependent(Y) and independent(X) variables
X = dfFootball.loc[:,['age','award']]
Y = dfFootball.loc[:,['current_value']]
```

```python
# Fit linear regression model - aka train model
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X, Y)
```

Fitting the model means that the linear regression algorithm will find the best-fitting line (or hyperplane in the case of multiple features) through the data points.

```python
# Display regression model
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

### 1. EDA in Python

##### ***Histogram of current value amounts where the bins cover $100000 increments***

This histogram will show the distribution (frequency, count) of the current values of football players with each bin having a width of $100. **Each bin represents a continuous range of values, and the height of the bin in the histogram represents the count (or frequency) of data points that fall within that range.**

By taking a smaller sample, you can perform exploratory data analysis (EDA) more efficiently while still obtaining meaningful insights. Using frac=0.1 will select 100 random rows from the original dataset.



```python
# Use a smaller sample of the dataset
df_sample = dfFootball.sample(frac=0.1, random_state=1)  # Adjust the fraction as needed

# Select relevant columns
df_reduced = df_sample[['current_value', 'age', 'goals', 'assists', 'award', 'position', 'team']]
```

```python
# Convert data types to more memory-efficient formats
df_reduced['current_value'] = pd.to_numeric(df_reduced['current_value'], downcast='float')
df_reduced['age'] = pd.to_numeric(df_reduced['age'], downcast='integer')
df_reduced['goals'] = pd.to_numeric(df_reduced['goals'], downcast='integer')
df_reduced['assists'] = pd.to_numeric(df_reduced['assists'], downcast='integer')
df_reduced['award'] = pd.to_numeric(df_reduced['award'], downcast='integer')

```

```python
# Set data fields
y = df_reduced['current_value']

# Set histogram parameters
binwidth = 100000 # Adjust bin width as needed for better visualization
binrange = np.arange(min(y), max(y) + binwidth, binwidth)

# Show plot
plt.figure(figsize=(10, 6))
plt.hist(y, bins=binrange, edgecolor='black')
plt.xlabel('Current Value')
plt.ylabel('Count')
plt.title('Histogram of Current Value')
plt.show()
```

As seen in the histogram, current values are primarily concentrated between 0 and 1 million dolars. The number of players with values above 1 million dolars decreases rapidly.

##### ***Scatterplot of Current Values vs. Awards***

```python
# Set data fields
x = dfFootball.loc[:, 'award']
y = dfFootball.loc[:, 'current_value']
```

```python
# Generate trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
```

```python
# Show plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data points')
plt.plot(x, p(x), "r--", label='Trendline')
plt.xlim(0, np.max(dfFootball.loc[:, 'award']))
plt.ylim(0, np.max(dfFootball.loc[:, 'current_value']))
plt.xlabel('Awards')
plt.ylabel('Current Value')
plt.title('Awards vs. Current Value with Trendline')
plt.legend()
plt.show()
```

Each blue dot represents a football player. Red line is a linear approximation of the relationship between awards and current value. There are many players with a lower number of awards. There are some outliers, for example left upper side we see players with high current values but fewer awards **indicating that other factors might also significantly influence a player's value.** There is a positive correlation between awards and value but relationshio is not perfectly linear and influenced by other factors.

##### ***Scatterplot of Current Value vs. Awards seperated by position and team factors***

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

def plot_scatter_with_trendline(ax, x, y, title):
    if not x.empty and not y.empty:
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        ax.scatter(x, y)
        ax.plot(x, p(x), "r--")
        ax.set_xlim(0, np.max(dfFootball['award']))
        ax.set_ylim(0, np.max(dfFootball['current_value']))
    ax.set_title(title)

## Top left plot
x = dfFootball.loc[(dfFootball['winger'] == 0) & (dfFootball['team'] == 'Leicester City'), 'award']
y = dfFootball.loc[(dfFootball['winger'] == 0) & (dfFootball['team'] == 'Leicester City'), 'current_value']
plot_scatter_with_trendline(axs[0, 0], x, y, 'Winger = No, Team = Leicester City')

## Top right plot
x = dfFootball.loc[(dfFootball['winger'] == 1) & (dfFootball['team'] == 'Leicester City'), 'award']
y = dfFootball.loc[(dfFootball['winger'] == 1) & (dfFootball['team'] == 'Leicester City'), 'current_value']
plot_scatter_with_trendline(axs[0, 1], x, y, 'Winger = Yes, Team = Leicester City')

## Bottom left plot
x = dfFootball.loc[(dfFootball['winger'] == 0) & (dfFootball['team'] == 'Manchester United'), 'award']
y = dfFootball.loc[(dfFootball['winger'] == 0) & (dfFootball['team'] == 'Manchester United'), 'current_value']
plot_scatter_with_trendline(axs[1, 0], x, y, 'Winger = No, Team = Manchester United')

## Bottom right plot
x = dfFootball.loc[(dfFootball['winger'] == 1) & (dfFootball['team'] == 'Manchester United'), 'award']
y = dfFootball.loc[(dfFootball['winger'] == 1) & (dfFootball['team'] == 'Manchester United'), 'current_value']
plot_scatter_with_trendline(axs[1, 1], x, y, 'Winger = Yes, Team = Manchester United')

plt.tight_layout()
plt.show()
```

**For the top left plot:**
*Dots show players who are not wingers in Leicester City. They are more clustered mostly with low award values. Trendline shows a negative slope which indicates that for non-winger players in Leicester City as the number of awards increases, their current value tends to decrease slightly. It might suggest that other factors contribute more significantly to the current value.*

**For the top right plot:**
*Fewer data points mean smaller sample size for wingers in this team.*

**For the bottom left plot:**
*Nonwinger players in Manchester City. Broader range of award values and wider spread in values. Outliers with high current values indicate variability. Positive slope indicates for nonwinger players in Manchester United number of awards is associated with an increase in current value. Awards appear to play more significant role.*

**For the bottom right plot:**
*Several data points are tightly clustered with low award values but show significant spread in value. For wingers in Manchester United, awards are a strong indicator of their current value. *


##### ***Scatterplot of Current Value vs. Minutes Played seperated by Position Encoded and Number of Games Injured factors***

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

def plot_scatter_with_trendline(ax, x, y, title):
    if not x.empty and not y.empty:
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        ax.scatter(x, y)
        ax.plot(x, p(x), "r--")
        ax.set_xlim(0, np.max(dfFootball['minutes played']))
        ax.set_ylim(0, np.max(dfFootball['current_value']))
    ax.set_title(title)

## Top left plot
x = dfFootball.loc[(dfFootball['games_injured'] >= 15) & (dfFootball['position_encoded'] == 4), 'minutes played']
y = dfFootball.loc[(dfFootball['games_injured'] >= 15) & (dfFootball['position_encoded'] == 4), 'current_value']
plot_scatter_with_trendline(axs[0, 0], x, y, 'Games Injured >= 15, Position Encoded = 4')

## Top right plot
x = dfFootball.loc[(dfFootball['games_injured'] < 15) & (dfFootball['position_encoded'] == 4), 'minutes played']
y = dfFootball.loc[(dfFootball['games_injured'] < 15) & (dfFootball['position_encoded'] == 4), 'current_value']
plot_scatter_with_trendline(axs[0, 1], x, y, 'Games Injured < 15, Position Encoded = 4')

## Bottom left plot
x = dfFootball.loc[(dfFootball['games_injured'] >= 15) & (dfFootball['position_encoded'] == 2), 'minutes played']
y = dfFootball.loc[(dfFootball['games_injured'] >= 15) & (dfFootball['position_encoded'] == 2), 'current_value']
plot_scatter_with_trendline(axs[1, 0], x, y, 'Games Injured >= 15, Position Encoded = 2')

## Bottom right plot
x = dfFootball.loc[(dfFootball['games_injured'] < 15) & (dfFootball['position_encoded'] == 2), 'minutes played']
y = dfFootball.loc[(dfFootball['games_injured'] < 15) & (dfFootball['position_encoded'] == 2), 'current_value']
plot_scatter_with_trendline(axs[1, 1], x, y, 'Games Injured < 15, Position Encoded = 2')

plt.tight_layout()
plt.show()
```

*In general there is a slight positive correlation between minutes played and current value, suggesting that players in this group who play more minutes tend to have higher current values. Players with fewer injuries generally show a stronger positive correlation between minutes played and current value compared to those with more injuries. Position 4 players benefit more from additional playing time in terms of market value compared to position 2 players. The positive correlation between minutes played and current value is weaker for players in position 2.*
